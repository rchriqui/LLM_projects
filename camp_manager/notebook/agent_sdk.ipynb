```python
import asyncio

from agents import Agent, Runner, function_tool
from dotenv import load_dotenv

from tool_schemas import (
    cancel_registration,
    get_camps,
    get_kids,
    get_registrations,
    register_kid,
    update_registration_status,
)

load_dotenv()

SYSTEM_PROMPT = """
You are a friendly and knowledgeable Summer Camp Registration Assistant.
You help parents register their children for summer camps, check availability,
manage existing registrations, and answer questions about camp schedules and pricing.

Guidelines:
- Always confirm with the user before creating, cancelling, or modifying any registration.
- If a name is ambiguous (e.g. multiple children named "Emma"), ask the parent to clarify.
- When a camp is full, proactively offer to add the child to the waitlist.
- Explain any validation errors (age restriction, schedule conflict, etc.) clearly and suggest alternatives.
- Keep responses concise and conversational.
"""

agent = Agent(
    name="Camp Registration Assistant",
    instructions=SYSTEM_PROMPT,
    tools=[
        function_tool(get_camps),
        function_tool(get_kids),
        function_tool(get_registrations),
        function_tool(register_kid),
        function_tool(cancel_registration),
        function_tool(update_registration_status),
    ],
)


async def main():
    result = await Runner.run(
        agent,
        "What camps are available for a 10-year-old?",
    )
    print(result.final_output)

```
